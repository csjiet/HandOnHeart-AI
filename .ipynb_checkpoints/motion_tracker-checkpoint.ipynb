## Tracking motion of 1 left hand landmark (left index finger)

```python
def filter_dataframe_by_landmarks(df, selected_landmarks, hand_prefix="rh_"):
    retained_columns = [f"{hand_prefix}{landmark.name}" for landmark, keep in selected_landmarks.items() if keep]
    
    if "frame" in df.columns:
        retained_columns.append("frame")
    filtered_df = df[retained_columns].copy()
    
    return filtered_df

selected_landmarks = {
    ht.CopyHandLandmark.WRIST: False,
    ht.CopyHandLandmark.THUMB_CMC: False,
    ht.CopyHandLandmark.THUMB_MCP: False,
    ht.CopyHandLandmark.THUMB_IP: False,
    ht.CopyHandLandmark.THUMB_TIP: False, 
    ht.CopyHandLandmark.INDEX_FINGER_MCP: False,
    ht.CopyHandLandmark.INDEX_FINGER_PIP: False,
    ht.CopyHandLandmark.INDEX_FINGER_DIP: False,
    
    ht.CopyHandLandmark.INDEX_FINGER_TIP: True, 
    
    ht.CopyHandLandmark.MIDDLE_FINGER_MCP: False,
    ht.CopyHandLandmark.MIDDLE_FINGER_PIP: False,
    ht.CopyHandLandmark.MIDDLE_FINGER_DIP: False,
    ht.CopyHandLandmark.MIDDLE_FINGER_TIP: False, 
    ht.CopyHandLandmark.RING_FINGER_MCP: False,
    ht.CopyHandLandmark.RING_FINGER_PIP: False,
    ht.CopyHandLandmark.RING_FINGER_DIP: False,
    ht.CopyHandLandmark.RING_FINGER_TIP: False, 
    ht.CopyHandLandmark.PINKY_MCP: False,
    ht.CopyHandLandmark.PINKY_PIP: False,
    ht.CopyHandLandmark.PINKY_DIP: False,
    ht.CopyHandLandmark.PINKY_TIP: False 
}
filtered_df = filter_dataframe_by_landmarks(df, selected_landmarks, hand_prefix="lh_")
print(f"Columns: {filtered_df.columns}")
```

```python
ht.plot_hand_motion(filtered_df.copy()) # left index finger
```
